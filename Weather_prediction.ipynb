Weather Prediction Using NaiveBayes Algorithm

```python
#Weather Prediction Using NaiveBayes Algorithm

import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
#Reading CSV files
df = pd.read_csv("/content/drive/MyDrive/weather_prediction_dataset.csv")
df
```

```python
#Encoding the strings to Numericals
Numerics = LabelEncoder()

#Dropping the target variable and make it as a new frame
inputs=df.drop('Play',axis='columns')
target=df['Play']
target
```

```python
#Create the new dataframe
inputs['outlook_n']= Numerics.fit_transform(inputs['Outlook'])
inputs['Temp_n']= Numerics.fit_transform(inputs['Temp'])
inputs['Hum_n']= Numerics.fit_transform(inputs['Humidity'])
inputs['win_n']= Numerics.fit_transform(inputs['Wind'])
inputs
```

```python
#Dropping the string values
inputs_n=inputs.drop(['Outlook','Temp','Humidity','Wind'],axis='columns')
inputs_n
```

```python
#Applying GaussianNB
classifier = GaussianNB()
classifier.fit(inputs_n,target)
```

```python
#Pridicting the accuracy
classifier.score(inputs_n,target)
```

```python
#Prediction
classifier.predict([[1,0,0,1]])
```
